# Using Imported Data

```python
import pandas as pd
data = pd.read_csv('data_bar.csv')
data
```

```python
from collections import Counter
```

```python
c = Counter(['python','js','C','C++','js','python','js'])
c
```

# Use our data

```python
ids = data['Responder_id']
lang_response = data['LanguagesWorkedWith']

language_counter = Counter()
```

```python
for response in lang_response:
    language_counter.update(response.split(';'))
```

```python
print(language_counter)
```

```python
print(language_counter.most_common(15))
```

```python
language = []
popularity = []

for item in language_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
```

```python
print(language)
```

```python
print(popularity)
```

## DataViz

```python
plt.figure(figsize=(20,5))

plt.bar(language, popularity)

plt.title('Most Popular Language')
plt.xlabel('Language')
plt.ylabel('Users')

plt.show()
```

```python
plt.barh(language, popularity)

plt.title('Most Popular Language')
plt.xlabel('Language')
plt.ylabel('Users')

plt.show()
```

```python
popularity.reverse()
language.reverse()
plt.barh(language, popularity)

plt.title('Most Popular Language')
plt.xlabel('Users')
plt.ylabel('Language')

plt.show()
```

# Materi Subplot

```python
plt.style.use('seaborn')
```

```python
data = pd.read_csv('data_sub.csv')
data
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
fig, ax = plt.subplots()
ax.plot(ages,py_salaries,label="Python")
ax.plot(ages,js_salaries, label="JavaScript", color="m")
ax.plot(ages, dev_salaries, label="All Dev", linestyle="--")

ax.legend()
ax.set_title("Median Salary (USD) vs Age")
ax.set_xlabel('Age')
ax.set_ylabel('Median Salary(USD)')

plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1)
ax1.plot(ages,py_salaries,label="Python")
ax2.plot(ages,js_salaries, label="JavaScript", color="m")
ax2.plot(ages, dev_salaries, label="All Dev", linestyle="--")

ax1.legend()
ax1.set_title("Median Salary (USD) vs Age")
ax1.set_xlabel('Age')
ax1.set_ylabel('Median Salary(USD)')

ax2.legend()
ax2.set_title("Median Salary (USD) vs Age")
ax2.set_xlabel('Age')
ax2.set_ylabel('Median Salary(USD)')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1, sharex = True)
ax1.plot(ages,py_salaries,label="Python")
ax2.plot(ages,js_salaries, label="JavaScript", color="m")
ax2.plot(ages, dev_salaries, label="All Dev", linestyle="--")

ax1.legend()
ax1.set_title("Median Salary (USD) vs Age")
ax1.set_xlabel('Age')
ax1.set_ylabel('Median Salary(USD)')

ax2.legend()
ax2.set_title("Median Salary (USD) vs Age")
ax2.set_xlabel('Age')
ax2.set_ylabel('Median Salary(USD)')

plt.tight_layout()
plt.show()
```

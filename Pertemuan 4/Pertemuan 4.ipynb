## Replace Missing Value

```python
df3 = new_df.replace({
    '?' : np.nan, 
    '--' : np.nan,
    'x' : np.nan,
    'y' : np.nan,
    'r' : np.nan,
    'a' : np.nan
})

df3
```

## DROP MISSING VALUE

```python
df3
```

```python
df3.dropna(axis=0, how='all', inplace=True)
df3
```

```python
df3.dropna(axis=0, how='any')
df3
```

```python
df3.dropna(axis=0, thresh=3)
```

## REPLACING MISSING VALUE

```python
df5 = df3.fillna(0)
df5
```

```python
df3
```

```python
df3.iloc[:,0] = df3.iloc[:,0].fillna(df3.iloc[:,0].mean())
df3
```

```python
df3.iloc[:,1] = df3.iloc[:,1].fillna(df3.iloc[:,1].median())
df3
```

```python
df3 = df3.drop(columns='col_c')
df3
```
